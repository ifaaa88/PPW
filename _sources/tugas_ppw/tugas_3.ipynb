# Penjelasan Metode Logistic Regression

 logistik regresi statistik digunakan, terutama untuk masalah klasifikasi biner, yaitu untuk memprediksi salah satu dari dua hasil yang mungkin (misalnya, "ya" atau "tidak", "1" atau "0"). Model ini memprediksi variabel kategoris, bukan variabel kontinu seperti regresi linear, meskipun namanya mengandung kata "regresi".
 Metode regresi logistik memprediksi apakah sesuatu termasuk dalam dua pilihan biner, seperti "ya" atau "tidak", "benar" atau "salah", atau hanya "1" dan "0".
Untuk melakukan ini, logistic regresi menghitung kemungkinan, atau kemungkinan bahwa sesuatu terjadi atau benar. Nilai kemungkinan ini akan selalu berkisar antara 0 dan 1. Sebagai contoh, jika kemungkinan adalah 0,8, itu berarti ada 80 persen kemungkinan hal itu akan terjadi, dan jika kemungkinan adalah 0,2, itu berarti ada 20 persen kemungkinan hal itu akan terjadi.

**menghitung nilai TF-IDF**

Kode ini digunakan untuk menghitung nilai TF-IDF dari teks yang terdapat di kolom stopword_removal dalam DataFrame df. TF-IDF (Term Frequency - Inverse Document Frequency) adalah teknik yang digunakan dalam text mining untuk mengukur seberapa penting suatu kata dalam suatu dokumen relatif terhadap seluruh koleksi dokumen.



- TfidfVectorizer diimpor dari library scikit-learn, yang digunakan untuk mengubah teks menjadi representasi numerik menggunakan teknik TF-IDF.
Frekuensi kata (TF): mengukur seberapa sering kata muncul dalam sebuah dokumen; Frekuensi dokumen balik (IDF): mengukur seberapa umum atau langka kata tersebut di seluruh dokumen.
Baris ini membuat objek TfidfVectorizer(), yang akan digunakan untuk menghitung nilai TF-IDF dari teks.
- TfidfVectorizer() memiliki banyak parameter yang dapat disesuaikan, tetapi dalam kode ini kita menggunakan pengaturan standar, yang artinya: Semua kata akan diperhitungkan (kecuali stopwords jika disediakan).
Semua kata akan diubah menjadi angka berdasarkan frekuensinya di dokumen dan seberapa umum kata tersebut di seluruh dokumen.
- Fit_transform() melakukan dua hal: fit(): Mempelajari kata-kata unik di kolom stopword_removal (yaitu, memetakan setiap kata ke nilai tertentu berdasarkan frekuensi dan IDF). transform(): Menghitung nilai TF-IDF untuk setiap kata dalam setiap dokumen dan mengubahnya menjadi bentuk matriks sparse (bentuk yang baik untuk menyimpan banyak nilai nol).
df['stopword_removal'] adalah kolom dalam DataFrame df yang berisi teks yang telah dibersihkan oleh proses penghapusan kata-kata umum seperti "dan", "di", dll.

tfidf_matrix adalah hasil dari fit_transform(), yang merupakan matriks yang berisi nilai TF-IDF dari setiap kata yang ada di setiap dokumen. Setiap baris dari matriks ini mewakili satu dokumen, dan setiap kolom mewakili satu kata. Matriks ini biasanya sangat besar dan sparce, dengan sebagian besar elemennya bernilai nol.
Tujuan Kode: Teknik TF-IDF digunakan untuk mengubah teks dalam kolom stopword_removal menjadi representasi numerik berdasarkan bobot penting setiap kata. Bobot ini dihitung berdasarkan frekuensi setiap kata muncul dalam dokumen dan seluruh kumpulan dokumen, sehingga model pembelajaran mesin dapat lebih mudah memahami data teks.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Menginisialisasi TfidfVectorizer
vectorizer = TfidfVectorizer()

# Menghitung TF-IDF
tfidf_matrix = vectorizer.fit_transform(df['stopword_removal'])
```

```python
# Menggunakan kolom 'cleaned_text' sebagai teks input dan 'kategori' sebagai label
texts = df['stopword_removal'].fillna('')  # Mengganti NaN dengan string kosong jika ada
labels = df['kategori'].fillna('unknown')  # Mengganti NaN di kategori jika ada

# Melihat label yang digunakan untuk klasifikasi
print(labels.unique())
```

**mengubah TF-IDF menjadi dataframe**

- menggunakan tfidf_matrix.toarray():
tfidf_matrix adalah hasil dari perhitungan TF-IDF sebelumnya yang berbentuk matriks sparse, yang hanya menyimpan nilai-nilai bukan nol untuk menghemat memori. toarray() mengubah matriks sparse menjadi matriks biasa, atau matriks penuh, sehingga kita dapat menggunakannya untuk membuat DataFrame Pandas. Hasilnya adalah matriks dengan nilai TF-IDF untuk setiap kata yang ditemukan di setiap dokumen.

- columns=vectorizer.get_feature_names_out():
Dengan menggunakan vectorizer.get_feature_names_out(), setiap kata unik yang ditemukan oleh TfidfVectorizer dari teks dikumpulkan. Setiap kata dalam teks akan memiliki nama kolom di DataFrame yang dibuat, dan setiap kolom akan mengandung nilai TF-IDF untuk kata tersebut di dokumen yang bersangkutan.

- menggunakan pd.DataFrame(...):
DataFrame dibuat dari array tfidf_matrix, di mana setiap baris menunjukkan sebuah dokumen, dan setiap kolom menunjukkan kata khusus dari kumpulan dokumen tersebut.
tfidf_df akan menyimpan nilai-nilai TF-IDF untuk setiap kata di setiap dokumen sehingga lebih mudah dibaca dan digunakan.

```python
# Mengubah hasilnya menjadi DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
tfidf_df.head(10)
```

```python
from sklearn.model_selection import train_test_split
# Melakukan transformasi TF-IDF
tfidf = TfidfVectorizer(max_features=10000)  # Menggunakan 5000 fitur teratas
X_tfidf = tfidf.fit_transform(texts)

# Memisahkan dataset menjadi training dan testing
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, labels, test_size=0.2, random_state=42)

# Melihat bentuk dari hasil TF-IDF
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

# Membuat model Logistic Regression dengan regularisasi
model = LogisticRegression(max_iter=1000, C=0.2)  # Mengurangi nilai C untuk menambah regularisasi

# Melatih model dengan data yang sudah diresample (jika diperlukan)
model.fit(X_train, y_train)

# Memprediksi hasil pada data testing
y_pred = model.predict(X_test)

```

```python
# Evaluasi hasil prediksi
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')

# Menampilkan confusion matrix dan laporan klasifikasi
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Alternatif: Menggunakan cross-validation untuk mengecek overfitting
from sklearn.model_selection import cross_val_score

# Menggunakan cross-validation
cv_scores = cross_val_score(model, X_train, y_train, cv=5)
print(f'Cross-validation accuracy: {cv_scores.mean() * 100:.2f}%')
```
